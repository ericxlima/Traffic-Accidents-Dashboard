```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('original_data_from_2020.csv', sep=';')

df.columns
```

```python
accidents_df = df.loc[(df['situacao'] == 'FINALIZADA')]
accidents_df
```

```python
vehicles_df = accidents_df[['auto', 'moto', 'ciclom', 'ciclista', 'onibus', 
                  'caminhao', 'viatura']]

vehicles_series = vehicles_df.sum(axis=1)


accidents_df = accidents_df[['data', 'natureza_acidente', 'bairro', 
                             'conservacao_via', 'pedestre']]
accidents_df['veiculos'] = vehicles_series
accidents_df
```

```python
accidents_df = accidents_df.loc[
    (accidents_df['conservacao_via'].notnull()) &
    (accidents_df['bairro'].notnull())
]

accidents_df
```

```python
accidents_df.replace(np.nan, 0, inplace=True)
accidents_df
```

```python
accidents_df.info()
```

```python
accidents_df = accidents_df.astype({'pedestre': int, 'veiculos':int})
accidents_df
```

```python
accidents_df.to_csv('processed_acidentes_2020.csv', index=False)
```
